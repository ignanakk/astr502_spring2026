```python
from astropy.io import fits
import matplotlib.pyplot as plt
import numpy as np
```

```python
file= fits.open('wasp136.fits')
```

```python
data= file[1].data
header=file[1].header
header
```

```python
print(data.columns)
```

```python
wave= data['WAVE'][0]
flux= data['FLUX'][0]
```

```python
print(wave)
```

```python
plt.plot(wave,flux)
plt.xlabel('wavelengths')
plt.ylabel('flux')
plt.title('WASP136 Spectra')
plt.show()
#plt.xlim(6500,6600) #hydrogen
#plt.xlim(6700,6720) #lithium
```

```python
plt.plot(wave,flux)
plt.xlabel('wavelengths')
plt.ylabel('flux')
plt.title('WASP136 Spectra, H-alpha')
plt.xlim(6500,6600) #hydrogen
plt.show()
```
